```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.DataFrame({"Salary":[70000,60000,52000,45000]})
df
```

```python
# MinMaxScaler

scaler = MinMaxScaler()
```

```python
scaler.fit_transform(df[["Salary"]])
```

```python
#add a col, of scaled values, next to salary
df['MinMax_wala'] = scaler.fit_transform(df[["Salary"]])
df
```

```python
scaler = StandardScaler()
```

```python
df["ss_wala"] = scaler.fit_transform(df[['Salary']])
df
```
